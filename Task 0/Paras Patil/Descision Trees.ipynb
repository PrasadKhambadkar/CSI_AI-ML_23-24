```python
import pandas as pd
df = pd.read_csv("/content/database.csv")
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df['Survived']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_PassengerId	= LabelEncoder()
le_Name	= LabelEncoder()
le_Sex = LabelEncoder()
```

```python
inputs['PassengerId_n'] = le_PassengerId.fit_transform(inputs['PassengerId'])
inputs['Name_n'] = le_PassengerId.fit_transform(inputs['Name'])
inputs['Sex_n'] = le_PassengerId.fit_transform(inputs['Sex'])
inputs.head()
```

```python
inputs_n = inputs.drop(['PassengerId','Name','Sex'],axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
 model = tree.DecisionTreeClassifier()
```

```python
 model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
pt = final_df.pivot_table(index='appid', columns='user_id', values='rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(appid):
    index = np.where(pt.index == appid)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:5]
    data = {}
    for i, score in similar_items:
        temp_df = final_df[final_df['appid'] == pt.index[i]]
        appid_key = temp_df.drop_duplicates('appid')['appid'].values[0]
        data[appid_key] = list(temp_df.drop_duplicates('appid')['appid'].values)
    return data
```

```python
recommendations = recommend(40)
```

```python
import ast

def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (SyntaxError, ValueError):
        return None

content_df['categories'] = content_df['categories'].fillna('').apply(safe_literal_eval)
content_df['genres'] = content_df['genres'].fillna('').apply(safe_literal_eval)
content_df['steamspy_tags'] = content_df['steamspy_tags'].fillna('').apply(safe_literal_eval)
```

```python
content_df['categories'] = content_df['categories'].apply(lambda x: [item.replace(';', ', ') if isinstance(item, str) else item for item in x] if x is not None else None)
content_df['genres'] = content_df['genres'].apply(lambda x: [item.replace(';', ', ') if isinstance(item, str) else item for item in x] if isinstance(x, list) else None)
content_df['steamspy_tags'] = content_df['steamspy_tags'].apply(lambda x: [item.replace(';', ', ') if isinstance(item, str) else item for item in (x if isinstance(x, list) else [])] if x is not None else None)
```

```python
content_df['short_description'] = content_df['short_description'].apply(lambda x: x.split() if isinstance(x, str) else [])
```

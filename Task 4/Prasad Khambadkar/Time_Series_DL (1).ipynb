## Removing all unnecessary columns

```python
def retaining_only_necessary_columns(dataframes, col1, col2):
    for i, df in enumerate(dataframes):
        df.columns = df.columns.str.strip()
        if col1 in df.columns and col2 in df.columns:
            df.drop(df.columns.difference([col1, col2]), axis=1, inplace=True)
```

### Calculating missing values in a dataframe and converting it into a dataframe

```python
def find_missing_values_summary(dfs):
    missing_values_summary = {}
    for idx, df in enumerate(dfs):
        for col in df.columns:
            if col not in missing_values_summary:
                missing_values_summary[col] = [0] * len(dfs)
            missing_values_summary[col][idx] = df[col].isnull().sum()

    missing_values_df = pd.DataFrame(missing_values_summary)
    missing_values_df.insert(0, 'DataFrame', [f'df{i + 1}' for i in range(len(dfs))])

    return missing_values_df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
first_game_data = result_array[0, 1, :].astype(float)

scaler = MinMaxScaler()
time_series_data = scaler.fit_transform(first_game_data.reshape(-1, 1)).flatten()

train_data = time_series_data[:6]
test_data = time_series_data[6:]

def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i+sequence_length]
        label = data[i+sequence_length]
        sequences.append((seq, label))
    return np.array(sequences)

sequence_length = 3
train_sequences = create_sequences(train_data, sequence_length)
test_sequences = create_sequences(test_data, sequence_length)

X_train, y_train = np.array([seq[0] for seq in train_sequences]), np.array([seq[1] for seq in train_sequences])
X_test, y_test = np.array([seq[0] for seq in test_sequences]), np.array([seq[1] for seq in test_sequences])

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], sequence_length, 1))

model = Sequential()
model.add(LSTM(50, input_shape=(X_train.shape[1], 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

y_pred = model.predict(X_test)

y_pred_denormalized = scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()

plt.plot(time_series_data, label='Actual Data')
plt.plot(np.arange(6, len(time_series_data)), y_pred_denormalized, label='Predicted Data')
plt.title('Time Series Forecasting for Hours Played dfs[0][0]')
plt.xlabel('Days')
plt.ylabel('Hours Played')
plt.legend()
plt.show()
```

## Calculating MinMaxScalar of each game's Hours Played in period of 9 days

```python
def calculate_feature_range_for_common_games(dataframes, feature_column, common_names):
    feature_ranges = {}

    for df in dataframes:
        for name, group in df.groupby('name'):
            if name in common_names:
                feature_values = group[feature_column].values

                min_value = min(feature_values)
                max_value = max(feature_values)

                if name not in feature_ranges:
                    feature_ranges[name] = {'min': min_value, 'max': max_value}
                else:
                    feature_ranges[name]['min'] = min(feature_ranges[name]['min'], min_value)
                    feature_ranges[name]['max'] = max(feature_ranges[name]['max'], max_value)

    return feature_ranges

feature_ranges_common_games = calculate_feature_range_for_common_games(dfs, 'Hours Played', common_names_array)
scalers_common_games = {name: MinMaxScaler(feature_range=(range_info['min'], range_info['max'])) for name, range_info in feature_ranges_common_games.items()}
```
